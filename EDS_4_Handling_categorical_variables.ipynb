**Handling categorical variables with one hot encoding**

Qualitative data :- Deals with descriptions (observable) - Categorical data

---


Quantity data :- Deals with numbers (measurable)



Machines don't understand categorical features/variables we need to convert categorical columns to numerical columns. This process is known as Categorical encoding.

**One-Hot encoding**


It is a process of creating dummy variables.

```python
import pandas as pd
#step 1 : Create the data
df = pd.DataFrame({"Team": ['A' , 'A' , 'B' , 'B' ,  'B' , 'B' ,  'C' , 'C'],'points':[25,12,15,14,19,23,25,29]})
```

```python
df
```

```python
#step 2 :-one hot encoding
#creating an instance of one - hot endoding
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder =  OneHotEncoder(handle_unknown = "ignore")
```

```python
#perform one hot encoding on 'team' column
encoder_df=pd.DataFrame(encoder.fit_transform(df[['Team']]).toarray())
```

```python
final_df=df.join(encoder_df)
```

```python
final_df
```

```python
#step 3 Drop the original categorical variables
#drop the team column
final_df.drop('Team',axis=1,inplace = True)
```

```python
#view final_df
final_df
```
